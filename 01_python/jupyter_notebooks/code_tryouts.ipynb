Data types means the kind of values these variables can hold. And, based on the datatypes different operations can be done on particular type of data.

## Integer

```python
var=10
print(var)
print(type(var))
```

## Float

```python
# decimal point tells it is a float
var2 = 12.33
print(var2)
print(type(var2))
```

## Complex Numbers

```python
x = 3 + 2j
print(x)
print(type(x))
```

## String

```python
string_var = "Hello World"
print(string_var)
print(type(string_var))
```

## Lists

```python
# list can hold any data types
# not memory efficient if we need faster computation
list1 = ["car", "bike", 3, 0.2]
print(list1)
print(type(list1))
```

## Tuple

```python
tup = ('car', 'bus', 'truck')
print(tup)
print(type(tup))
```

## Dictionary

```python
dict_one = {"Name":"Shym", "School":"Ohio U", "age":32}
print(dict_one)
print(type(dict_one))
```

## Set

```python
set_one = set({"hello", "world", "hello"})  # can't save duplicates
print(set_one)
print(type(set_one))
```

## Boolean

```python
a = True
print(a)
print(type(a))
```

```python
# create a list: can contain same data types of mixed types
list_one = ["sunday", "monday", "tuesday", "wednesday", "thursday", "friday", "saturday", 0, 12, 2.3]
print(list_one)
```

```python
# find len of list
len(list_one)
```

```python
# nested list
list_two = ["sunday", ["monday", "tuesday"], "wednesday", 0, 2, 12.2]
print(list_two)
```

```python
print(len(list_two))
```

```python
# slicing list: lists are 0-indexed
list_one[2:5]
```

```python
list_one[:]
```

```python
# first four items
list_one[:4]
```

```python
# last four
list_one[-4:]
```

```python
list_one.append("Jan")
print(list_one)
print(len(list_one))
```

```python
# updating value
list_one[1] = "Mon"
print(list_one)
```
